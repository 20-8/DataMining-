```python
dataframe = train_transaction["TransactionAmt"]
dataframe = pd.concat([dataframe,train_transaction["ProductCD"]],axis=1)
dataframe = pd.concat([dataframe,train_transaction["card1"]],axis=1)
dataframe = pd.concat([dataframe,train_transaction["card2"]],axis=1)
dataframe = pd.concat([dataframe,train_transaction["card4"]],axis=1)
dataframe = pd.concat([dataframe,train_transaction["card6"]],axis=1)
dataframe = pd.concat([dataframe,train_transaction["P_emaildomain"]],axis=1)
dataframe = pd.concat([dataframe,train_transaction["R_emaildomain"]],axis=1)
dataframe = pd.concat([dataframe,train_transaction["C1"]],axis=1)
dataframe = pd.concat([dataframe,train_transaction["C2"]],axis=1)
dataframe = pd.concat([dataframe,train_transaction["C3"]],axis=1)
dataframe = pd.concat([dataframe,train_transaction["C4"]],axis=1)
dataframe = pd.concat([dataframe,train_transaction["C5"]],axis=1)
dataframe = pd.concat([dataframe,train_transaction["C6"]],axis=1)
dataframe = pd.concat([dataframe,train_transaction["C7"]],axis=1)
dataframe = pd.concat([dataframe,train_transaction["C8"]],axis=1)
dataframe = pd.concat([dataframe,train_transaction["C9"]],axis=1)
dataframe = pd.concat([dataframe,train_transaction["V279"]],axis=1)
dataframe = pd.concat([dataframe,train_transaction["V280"]],axis=1)
dataframe = pd.concat([dataframe,train_transaction["V285"]],axis=1)
```

```python
type(dataframe["R_emaildomain"][0]) == type (np.nan)
```

```python
for i in range (dataframe.shape[1]):
    dataframe[dataframe.columns[i]] = dataframe[dataframe.columns[i]].fillna(value = np.unique(list(dataframe[dataframe.columns[i]]))[np.argsort(np.unique(list(dataframe[dataframe.columns[i]]),return_counts=True)[1])[-2]])
```

```python
#Extracting dates and times from the columns TransactionsDT
import datetime
#Via Kernels
START_DATE = '2017-12-01'
startdate = datetime.datetime.strptime(START_DATE, '%Y-%m-%d')
train_transaction['TransactionDT'] = train_transaction['TransactionDT'].apply(lambda x: (startdate + datetime.timedelta(seconds = x)))

train_transaction['year'] = train_transaction['TransactionDT'].dt.year
train_transaction['month'] = train_transaction['TransactionDT'].dt.month
train_transaction['dow'] = train_transaction['TransactionDT'].dt.dayofweek
train_transaction['hour'] = train_transaction['TransactionDT'].dt.hour
train_transaction['day'] = train_transaction['TransactionDT'].dt.day
```

```python
#Plotting No of Transactions vs Time(date)
time = ["year","month","dow","hour","day"]
for i in range (0,5):
    figure = plt.figure
    ax = plt.gca()
    ax.scatter(train_transaction["TransactionID"],train_transaction[time[i]] )
    ax.set_ylabel(time[i])
    ax.set_xlabel("TransactionID")
    ax.set_title("{} wrt {}".format("TransactionID", time[i]))

    plt.legend()
    plt.show()
```

```python
#PCA Analysis
from sklearn.decomposition import PCA
pca = PCA(n_components=2) #We have transformed from 20 attributes to 2
dataframe_PCA = pca.fit_transform(dataframe)
print("PCA",dataframe_PCA.shape)
rand= np.random.randint(0,2,dataframe.shape[0])
plt.scatter(dataframe_PCA[:,0],dataframe_PCA[:,1],cmap=rand)
```

```python
#LDA Analysis without PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
classifier = LDA()
classifier.fit(dataframe[:500000],train_transaction["isFraud"][:500000])
print(np.sum(classifier.predict(dataframe[500000:])==train_transaction["isFraud"][500000:])/train_transaction[:500000].shape[0])
```
